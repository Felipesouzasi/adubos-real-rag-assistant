```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.chat_models.openai import ChatOpenAI
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import PyPDFLoader
from langchain_classic.chains.question_answering import load_qa_chain
```

```python
#Load dos modelos (Embeddings e LLM)

embeddings_model = OpenAIEmbeddings()
llm_model = ChatOpenAI(model_name="gpt-3.5-turbo", max_tokens=200)
```

```python
# Carregar o PDF

pdf_link = "CATÁLOGO - SAFRA REAL 2026.pdf"

loader = PyPDFLoader(pdf_link, extract_images=False)
pages = loader.load_and_split()
```

```python
# Separar em Chunks (pedaços menores para facilitar o processamento)
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000, 
    chunk_overlap=20,
    length_function=len,
    add_start_index=True
)

chunks = text_splitter.split_documents(pages)
```

```python
# Salvar num Vecto DB (ChromaDB)

db = Chroma.from_documents(chunks, embedding=embeddings_model, persist_directory="text_index")
db.persist()
```

```python
# Carregar DB
vectordb = Chroma(persist_directory="text_index", embedding_function=embeddings_model)

# Load Retriever (buscar os chunks relevantes para a pergunta)
retriever = vectordb.as_retriever(search_kwargs={"k": 3})
chain = load_qa_chain(llm_model, chain_type="stuff")
```

```python
def ask_question(question):
    # Nas versões recentes do LangChain, 'get_relevant_documents' foi substituído por 'invoke'
    contexts = retriever.invoke(question)
    # A chamada da chain também mudou para o método 'invoke' passando um dicionário
    answer = chain.invoke({"input_documents": contexts, "question": question})['output_text']
    return answer
```

```python
import json
from langchain_core.documents import Document

# 1. Carregar o JSON gerado
json_link = "catalogo_consolidado.json"  # Nome do arquivo gerado pelo nosso script anterior
try:
    with open(json_link, 'r', encoding='utf-8') as f:
        json_data = json.load(f)
        
    documents = []
    # 2. Converter o JSON para o formato Document do LangChain
    for page in json_data.get("pages", []):
        if not page.get("text", "").strip():
            continue
        doc = Document(
            page_content=page.get("text", ""),
             metadata={
                "page": page.get("page", 0),
                "source": page.get("source", json_link),
                "catalog": page.get("catalog", ""),
            }
        )
        documents.append(doc)
        
    print(f"Foram carregadas {len(documents)} páginas do JSON com sucesso!")
    
    # 3. Separar em Chunks usando a mesma lógica do PDF
    chunks_json = text_splitter.split_documents(documents)
    print(f"Documento dividido em {len(chunks_json)} chunks.")
    
    # 4. Salvar num novo Vector DB (ChromaDB) específico para o JSON
    db_json = Chroma.from_documents(chunks_json, embedding=embeddings_model, persist_directory="text_json_index")
    db_json.persist()
    print("VectorDB do JSON criado e salvo!")
    
    # 5. Configurar o novo retriever e a nova chain
    retriever_json = db_json.as_retriever(search_kwargs={"k": 5})
    chain_json = load_qa_chain(llm_model, chain_type="stuff")
    
    def ask_question_json(question):
        contexts = retriever_json.invoke(question)
        answer = chain_json.invoke({"input_documents": contexts, "question": question})['output_text']
        return answer
        
    print("\nSistema RAG com JSON pronto para uso!")
    
except FileNotFoundError:
    print(f"Arquivo {json_link} não encontrado. Certifique-se de rodar o script convert_to_json.py antes!")
```

```python
# Testando a solução
question = "eu quero começar a plantar cana-de-açúcar, quais as recomendações?"
resposta = ask_question_json(question)

print(f"Pergunta: {question}")
print(f"Resposta:\n{resposta}")
```
